```python
# Indexing by position
s = pd.Series([1,3,5,7,9])
print(s.iloc[0])                  # iloc
print(s.iloc[-1])
```

```python
# Indexing by label                    # loc
print(s.loc[0]) # First element
print(s.loc[4]) # Last element
```

```python
# Create a Series
s = pd.Series([10,20,30,40], index=['a','b','c','d'])
print(s)
# Accessing elements using iloc
print(s.iloc[0])      # Access the first element
print(s.iloc[2])      # Access elements at positions 1 and 2 (exclusive) of 3

# Accessing elements using loc
print(s.loc['a'])     # Access the element with label 'a'
print(s.loc['b':'c'])     # Access the element with label 'b' and 'c'
```

```python
# Filtering elements greater than 20
filtered = s[s > 20]
print(filtered)
```

```python
print(s.sum())
```

```python
# Cumilative sum of the Series
print(s)
print(s.cumsum())
```

```python
# Aggregating using multiple operations
agg = s.agg(['sum','mean','std'])
print(agg)
```

```python
# Create a Data Frame
data = {'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]}
data = pd.DataFrame(data)
print(data)
```

```python
# Define a function to sum two series
def sum_series(x, y):
    return x + y

# Apply the function on multiple Series using apply()
result = data.apply(lambda x: sum_series(x['A'], x['B']), axis=1)
print(result)
```
