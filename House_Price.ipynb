# Data Cleaning

```python
# Remove non-numeric characters from 'Price' and convert to float
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Extract floor number from 'Floor'
df['Floor Number'] = df['Floor'].str.extract('(\d+)').astype(float)
```

```python
# Handle missing values and non-numeric values before converting to integers
df['Bathroom'] = pd.to_numeric(df['Bathroom'], errors='coerce').fillna(0).astype(int)
df['Balcony'] = pd.to_numeric(df['Balcony'], errors='coerce').fillna(0).astype(int)

df['Bathroom'].loc[df['Bathroom'] == '>10'] = 11
df['Balcony'].loc[df['Balcony'] == '>10'] = 11
```

```python
# mean imputation
df['Bathroom'] = df['Bathroom'].fillna(df['Bathroom'].mean())
df['Balcony'] = df['Balcony'].fillna(df['Balcony'].mean())
df['Price']  = df['Price'].fillna(df['Price'].mean())
```

```python
# filling null values with mode()
df['Furnishing'] = df['Furnishing'].fillna(df['Furnishing'].mode()[0])
df['Status']  =  df['Status'].fillna(df['Status'].mode()[0])
df['Transaction']  =  df['Transaction'].fillna(df['Transaction'].mode()[0])
```

```python
# Function to convert area to numeric
def convert_area(area):
    if isinstance(area, float) or pd.isna(area):  # Check if area is missing or a float
        return None
    if 'sqft' in area:
        return float(area.replace(' sqft', ''))
    elif 'sqm' in area:
        return float(area.replace(' sqm', '')) * 10.764  # Convert to square feet
    else:
        return None

# Apply conversion to 'Carpet Area' column
df['Carpet Area'] = df['Carpet Area'].apply(convert_area)
```

```python
df['Carpet Area'] = df['Carpet Area'].fillna(df['Carpet Area'].mean())
```

```python
df['Floor'] = df['Floor'].fillna(df['Floor'].mode()[0])
```

```python
# prompt: df['Amount'] converting Lac and Cr to the rupees

def convert_amount(amount):
  if isinstance(amount, float) or pd.isna(amount):
    return None
  if 'Lac' in amount:
    return float(amount.replace(' Lac', '')) * 100000
  elif 'Cr' in amount:
    return float(amount.replace(' Cr', '')) * 10000000
  else:
    return None

df['Amount'] = df['Amount'].apply(convert_amount)
```

# Data Splitting

```python
X = df.drop('Amount', axis=1)
Y = df['Amount']
```

```python
df.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

# Model Building

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(),
    'Decision Tree': DecisionTreeRegressor(),
    'Support Vector Machine': SVR(),
    'K-Nearest Neighbors': KNeighborsRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(random_state=42)
}
```

# Prediction

```python
model.values()
```

```python
# Evaluate models
results = []
for name, model in model.items(): # Use model.items() to iterate through key-value pairs
    model.fit(X_train, Y_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(Y_test, y_pred, squared=False))
    r2 = r2_score(Y_test, y_pred) # Changed variable name to avoid overwriting the function
    mse = mean_squared_error(Y_test, y_pred, squared=False)
    results.append({"Model": name,'MSE': mse, "r2_score": r2, "RMSE": rmse}) # Use r2 instead of r2_score

# Create a results DataFrame
results_df = pd.DataFrame(results)

# Print the results
print(results_df)
```

```python
# Plot the results
sns.barplot(x='Model', y='RMSE', data=results_df, palette='viridis')
plt.title('Model RMSE Comparison')
plt.ylabel('Root Mean Squared Error')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Model', y='MSE', data=results_df, palette='rainbow')
plt.title('Model MSE Comparison')
plt.ylabel('Mean Squared Error')
plt.xticks(rotation=45)
plt.show()
```
