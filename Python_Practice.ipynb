```python
def prime_finder(n):
  if n<=1:
    return 'not prime'
  elif n==2 or n==3:
    return 'prime'
  elif n>3:
    for i in range(2,int(n**0.5)+1):
      return 'not prime'
  else:
    return 'prime'
```

```python
prime_finder(12)
```
