```python
# keeping the projection = 3d
# creates the 3d plot
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ='3d')

# Creating plot
ax.scatter3D(df['Age'], df['Salary'], df['Years of Experience'],
             color = 'green')
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
ax.set_zlabel('Years of Experience')

# show plot
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="smoker", size="size") # Assuming you want to map the size to a variable called 'size'
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
import plotly.express as px

# Remove the 'style' argument
fig = px.scatter(tips, x="total_bill", y="tip", color="smoker", size="size")
fig.update_layout(
    xaxis_title="Total Bill",
    yaxis_title="Tip",
    title="Total Bill vs Tip"
)
fig.show()
```

```python
g = sns.relplot(data=tips, x="total_bill", y="tip", col="time", hue="smoker", style="smoker", size="size")
g.set_titles('Restaurant Tips')
plt.show()
```

```python
g = sns.FacetGrid(tips, col="day", hue='sex')
g.map(sns.histplot, "tip");
```

```python
g = sns.FacetGrid(tips, col="day", hue='sex')
g.map(sns.scatterplot, "total_bill", "tip");
```

```python
h = sns.displot(data=tips, x="total_bill", col="time", kde=True)
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',hue='day',data=tips,palette='rainbow')
```

```python
fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal",
             hue="event", style="region",
             data=fmri,markers=True,dashes=True)
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.title('Timepoint vs Signal')
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.barplot(x='pclass', y='fare', hue='sex', palette='bright', errorbar=("ci", 95), data=titanic)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Pclass vs Fare')
plt.show()
```

```python
sns.violinplot(x='pclass', y='fare', palette='bright', data=titanic)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Pclass vs Fare')
plt.show()
```

```python
sns.jointplot(x='age', y='fare', palette='bright', data=titanic)
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Pclass vs Fare')
plt.show()
```

```python
# load sample dataset
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='carat', y='price', data=diamonds)
# add annotations using plt.text instead of plt.txt
plt.text(0.5, 15000, 'Large, Expensive Diamonds', fontsize=18, color='black')
plt.text(2.5, 5000, 'Small, Affordable Diamonds', fontsize=18, color='black')
# customize plot
plt.xlabel('Carat (ct)')
plt.ylabel('Price ($)')
plt.title('Diamond Prices by Carat')
plt.show()
```
