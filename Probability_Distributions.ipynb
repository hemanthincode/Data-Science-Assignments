```python
# normal distribution n(mue, sigma)
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
x = random.normal(25, 3, 500)
fig = px.histogram(x)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

x = random.normal(25, 3, 500)

# Instead of go.Density, use go.Histogram with histnorm='probability density'
fig = go.Figure(data=[go.Histogram(x=x, histnorm='probability density')])

# Update layout for better visualization (optional)
fig.update_layout(title="Kernel Density Estimation Plot",
                  xaxis_title="Values",
                  yaxis_title="Density")

fig.show()
```

```python
z = random.normal(25, 3, 500)
sns.kdeplot(z)
print(z)
plt.show()
```

```python
import scipy.stats as stats

z = random.normal(25, 3, 500)

# Calculate mean, min, max, skewness, and kurtosis using appropriate functions
mean = z.mean()
minimum = z.min()
maximum = z.max()
skewness = stats.skew(z)  # Use scipy.stats.skew
kurtosis = stats.kurtosis(z)  # Use scipy.stats.kurtosis

print(f"Mean: {mean}, Min: {minimum}, Max: {maximum}, Skewness: {skewness}, Kurtosis: {kurtosis}")
```

```python
# comparing all the distribution with size = 5000, mean=50
sns.kdeplot(random.normal(50, 5, 5000), label='normal')
sns.kdeplot(random.binomial(100, 0.5, 5000), label='binomial')
sns.kdeplot(random.poisson(50, 5000), label='poisson')
plt.legend()
plt.show()
```

```python
# Take into account that 80% of customers visiting the museum store ultimately make a purchase.
# If a random sample of 10 customers is chosen for a day,
# What is the probability of a specific number of visitors purchasing a product ?
# Let's begin the calculation by defining the values.
```

```python
n = 10 # number of visitors selected randomly
p = 0.80 # the probability that a visitor will end up buying a souvenir
k = np.arange(0,11) # different possible number of visitors selected in a numpy array i.e. 0 or 5 etc
k
```

```python
# generate the probability distribution
binomial = stats.binom.pmf(k, n, p)
binomial
```

```python
sum(binomial)
```

```python
# style & colors
sns.set(style='white')
colors = sns.color_palette('viridis',n_colors=len(binomial))

# plotting the distribution with gradient colot bars
plt.figure(figure=(10,6))
sns.barplot(hue=k, y=binomial, palette=colors)

sns.despine(left=True, bottom=True)

# setting the title and labels
plt.title('Binomial Distribution: n=%i, p=%.2f' % (n, p), fontsize=18, color='darkgreen')
plt.xlabel('Number of Sucesses', fontsize=14, color='darkgreen')
plt.ylabel('Probability of Success', fontsize=14, color='darkgreen')
plt.xticks(color='darkgreen')
plt.yticks(color='darkgreen')

plt.show()
```

```python
# pdf and cdf for a normal distribution
from scipy.stats import norm
from matplotlib import pyplot as plt
# define distribution parameters
mu = 50
sigma = 5
# create distribution
dist = norm(mu, sigma)
# plot pdf
values = [value for value in range(30,70)]
probabilities = [dist.pdf(value) for value in values]
plt.plot(values, probabilities, 'b', label='pdf')
plt.legend()
plt.show()
# plot cdf
values = [value for value in range(30,70)]
probabilities = [dist.cdf(value) for value in values]
plt.plot(values, probabilities, 'r', label='cdf')
plt.legend()
plt.show()
```
