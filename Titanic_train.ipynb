```python
def impute_age(cols):
  Age = cols[0]
  Pclass = cols[1]

  if pd.isnull(Age):
    if Pclass == 1:
      return 37
    elif Pclass == 2:
      return 29
    else:
      return 24
  else:
    return Age
```

```python
df['Age'] = df[['Age','Pclass']].apply(impute_age, axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
pd.get_dummies(df['Embarked'],drop_first=True).head()
```

```python
sex = pd.get_dummies(df['Sex'],drop_first=True).head()
embark = pd.get_dummies(df['Embarked'],drop_first=True).head()
```

```python
df.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df =pd.concat([df,sex,embark],axis=1)
df.head()
```

```python
# prompt: Using dataframe df: do label encoding for male, q and s

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

df['male'] = label_encoder.fit_transform(df['male'])
df['Q'] = label_encoder.fit_transform(df['Q'])
df['S'] = label_encoder.fit_transform(df['S'])
```

```python
df.drop('Survived',axis=1).head()
```

```python
df['Survived'].head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Survived',axis=1),
                                                  df['Survived'],test_size=0.3,
                                                  random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
```

```python
logr.fit(x_train,y_train)
```

```python
predictions = logr.predict(x_test)
predictions
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
confusion_matrix = confusion_matrix(y_test,predictions)
accuracy = accuracy_score(y_test,predictions)
```
